# KCI (Korean)

```python
import os
import json
import pandas as pd

filepath = os.path.join('data', '20200109-21-40-45_preprocessed.json')
with open(filepath,"r") as f:
    d = json.load(f)
df = pd.read_json(d)
df.head()
```

```python
sentence_list = list()
for doc_id, row in df.iterrows():
    for sentence_id, sentence in enumerate(row['sentences']):
        sentence_list.append([doc_id, sentence_id, sentence])
result_df = pd.DataFrame(sentence_list, columns = ['document_id', 'sentence_id', 'sentence']) 
result_df
```

```python
result_df.to_csv(os.path.join('output', 'kci_sentences_%d.csv' % len(result_df) ), index=False)
```
